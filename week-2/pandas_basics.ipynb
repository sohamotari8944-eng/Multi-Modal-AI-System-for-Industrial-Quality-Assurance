# Introduction to Pandas

## What is Pandas?
Pandas is the most popular Python Data Analysis Library, used for manipulating, cleaning, and analyzing structured data.

## Why use Pandas?
Loads data fast: Reads CSV, Excel, and SQL files in a single line.
Uses vectorized math to process millions of rows instantly.
Drops or fills missing values and removes duplicates easily.
