## Decision Tree

```python
import pandas as pd
```

```python
df = pd.read_csv("company_salary.csv")
df
```

```python
inputs = df.drop('Salary_more_then_100k',axis=1)
target = df['Salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Company = LabelEncoder()
le_Job = LabelEncoder()
le_Degree = LabelEncoder()
```

Creating extra column in inputs

```python
inputs['Company_n'] = le_Company.fit_transform(inputs['Company'])
inputs['Job_n'] = le_Job.fit_transform(inputs['Job'])
inputs['Degree_n'] = le_Degree.fit_transform(inputs['Degree'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Company','Job','Degree'],axis=1)
```

```python
inputs_n
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs_n, target, test_size=0.4, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
target
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.fit(x_train,y_train)
```

The Score will be one becuase all the datamis trained and same is tested so it will be correct always.

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[2,0,0]])
```

```python
model.score(inputs_n,target)
```
